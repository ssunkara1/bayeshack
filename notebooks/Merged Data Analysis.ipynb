```python
from __future__ import print_function
import pandas as pd
```

```python
scores = pd.read_csv('./../analysis_data/scores_by_zipcodes.csv', header=None)
scores.columns = ['zip_code', 'scores']
preds = pd.read_csv('./../analysis_data/predictors.csv')

new_preds = preds.set_index('zip code')
scores = scores.set_index('zip_code')
final_data_frame = pd.concat([new_preds, scores], axis=1).dropna()
final_data_frame
```
